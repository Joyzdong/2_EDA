```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# plt.rcParams['font.family'] = 'Malgun Gothic'  # 폰트 적용 (한글 깨짐 방지)
plt.rcParams['font.size'] = 14                 # 폰트 크기
plt.rcParams['figure.figsize'] = (13, 8)       # figsize 를 전역으로 설정
plt.rcParams['axes.unicode_minus'] = False     # -(마이너스) 표기 가능하도록 함
```

```python
penguins = sns.load_dataset('penguins')
penguins = penguins.dropna()
penguins
```

```python
ax = sns.countplot(penguins, x='species', hue='sex', palette='Set3',
                   order=penguins['species'].value_counts().index)
# count 표시
for rec in ax.patches:
    height = int(rec.get_height())
    x, y = rec.get_x() + rec.get_width() / 2, height + 3
    ax.text(x, y, height, ha='center')
ax.set_ylim(top=ax.get_ylim()[1] + 10)
plt.show()
```

```python
sns.scatterplot(penguins, x='bill_depth_mm', y='bill_length_mm', hue='species', size='sex')
plt.show()
```
